```python
import pandas as pd
cafe_data = pd.read_csv('Database_CSVs/cafe_entity.csv')
```

```python
cafe_data
```

```python
keywords = ['starbucks', 'nero', 'costa', 'pret']
```

```python

cafe_data['Top_5'] = cafe_data['Cafe_Name'].apply(lambda x: any(word in x.lower() for word in keywords))
```

```python
Number_top_5 = cafe_data['Top_5'].sum()
print(Number_top_5)
```

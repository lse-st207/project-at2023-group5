```python
import pandas as pd
```

```python
main_cafe = pd.read_csv('cafe.csv')
```

```python
main_cafe.head()
```

```python
main_cafe['Cafe_Id'] = range(1, len(main_cafe) + 1)
main_cafe.head()
```

```python

cafe_entity = main_cafe[['Cafe_Id', 'Name','Types','lat','lng']].rename(columns={'Name': 'Cafe_Name', 'Types': 'Cafe_Types', 'lat': 'Latitude', 'lng': 'Longitude'})

cafe_entity.head()
```

```python
location_entity = main_cafe[['Cafe_Id', 'Vicinity','lat','lng']].rename(columns={'lat': 'Latitude', 'lng': 'Longitude'})

location_entity.head()
```

```python
rating_entity = main_cafe[['Cafe_Id', 'Rating','User Ratings Total']].rename(columns={'User Ratings Total': 'Number_Ratings'})

rating_entity



```
